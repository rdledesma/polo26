# Listas en Python


## 1. ¿Qué es una lista?


```python
numeros = [1, 2, 3, 4, 5]
nombres = ["Ana", "Luis", "Pedro"]
mezcla = [1, "hola", 3.14, True]

print(numeros)
print(nombres)
print(mezcla)
```

## 2. Acceso a elementos


```python
lista = [10, 20, 30, 40]

print(lista[0])
print(lista[-1])
```

## 3. Modificar listas


```python
lista = [1, 2, 3]
lista[1] = 99
print(lista)
```

## 4. Métodos importantes


```python
lista = [1, 2, 3]

lista.append(4)
lista.insert(1, 10)
lista.remove(2)
ultimo = lista.pop()

print(lista)
print("Elemento eliminado:", ultimo)
```

## 5. Recorrer listas


```python
for elemento in lista:
    print(elemento)
```

## 6. Funciones útiles


```python
nums = [4, 2, 8, 1]

print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
```

## 7. Slicing


```python
nums = [0,1,2,3,4,5]

print(nums[1:4])
print(nums[:3])
print(nums[::2])
```

## 8. Listas con condicionales


```python
nums = [1,2,3,4,5,6]

for n in nums:
    if n % 2 == 0:
        print(n)
```

## 9. Listas por comprensión


```python
cuadrados = [x**2 for x in range(10)]
pares = [x for x in range(20) if x % 2 == 0]

print(cuadrados)
print(pares)
```
