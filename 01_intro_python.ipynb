# Introducción a Python

Curso para principiantes

## ¿Qué es un lenguaje de programación?

Un lenguaje de programación es una forma de comunicarse con la computadora para indicarle qué hacer.

Permiten:
- Automatizar tareas
- Resolver problemas
- Procesar datos
- Crear software


## ¿Por qué existe Python?

Python fue creado para ser:
- Fácil de leer
- Fácil de aprender
- Potente

Se usa en:
- Ciencia de datos
- Inteligencia artificial
- Desarrollo web
- Automatización


## Primer programa: print()

```
print("Hola mundo")
```

## Entrada de datos: input()

```
nombre = input("¿Cómo te llamas? ")
print("Hola", nombre)
```

## Variables

```
edad = 30
altura = 1.75
nombre = "Ruben"

print(nombre)
print(edad)
print(altura)
```

## Tipos de datos básicos

- int: enteros
- float: números con decimales
- str: texto
- bool: verdadero o falso


```
x = 10
y = 3.14
texto = "Python"
es_valido = True

print(type(x))
print(type(y))
print(type(texto))
print(type(es_valido))
```
