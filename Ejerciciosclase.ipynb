```python
#entrada n valores ingresados por consola
#salida el mayor de los valores ingresados, asumiendo que todos los
#valores son mayores que 0

#ejemplo con n = 4
# n1 = 2
# n2 = 4
# n3 = 9
# n4 = 3
# salida 9
```

```python
n = int(input("ingrese n"))
menor = int(input("cargue el valor 0"))

j = 1
while(j < n-1):
    o = int(input(f"cargue el valor {j}"))
    if o < menor:
        menor = o
    j = j +1
print(f"el menor es {menor}")
```
