# Condicionales en Python (Extenso)

Guía completa para principiantes

## ¿Qué es un condicional?

Un condicional permite tomar decisiones en el programa.

Ejemplo:
- Si llueve → llevo paraguas
- Si no → no lo llevo


## Estructura básica: if

```
edad = 18

if edad >= 18:
    print("Eres mayor de edad")
```

## if - else

```
edad = 16

if edad >= 18:
    print("Mayor de edad")
else:
    print("Menor de edad")
```

## if - elif - else

```
nota = 7

if nota >= 9:
    print("Excelente")
elif nota >= 6:
    print("Aprobado")
else:
    print("Desaprobado")
```

## Operadores de comparación

```
x = 10
y = 5

print(x > y)
print(x < y)
print(x == y)
print(x != y)
```

## Operadores lógicos

```
edad = 20
tiene_dni = True

if edad >= 18 and tiene_dni:
    print("Puede votar")
```

## Variables simples en decisiones

```
temperatura = 30

if temperatura > 25:
    print("Hace calor")
else:
    print("Hace frío")
```

## Ejemplo con input

```
edad = int(input("Ingrese su edad: "))

if edad >= 18:
    print("Mayor de edad")
else:
    print("Menor de edad")
```
