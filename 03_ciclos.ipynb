Muchas veces necesitamos repetir acciones:

- Procesar muchos datos

Para eso usamos **ciclos**


# Ciclo `while`

Repite instrucciones **mientras una condición sea verdadera**.

## Sintaxis

```python
while condicion:
    # código
```


````python
i = 1

while i <= 5:
    print(i)
    i += 1
````

## Cuidado: loops infinitos

Si la condición nunca cambia → el programa no termina.


````python
# Ejemplo incorrecto (NO ejecutar indefinidamente)
i = 1
while i <= 5:
    print(i)
    # falta actualizar i
````

# Ciclo `for`

Se usa para recorrer secuencias o repetir una cantidad fija de veces.


````python
for i in range(5):
    print(i)
````

## Uso de range()

- `range(5)` → 0 a 4
- `range(1,6)` → 1 a 5
- `range(2,10,2)` → pares


````python
for i in range(2, 10, 2):
    print(i)
````

# Ejemplo: suma acumulada

````python
suma = 0
for i in range(1, 6):
    suma += i
print("Suma:", suma)
````

# Control de ciclos

## break → corta el ciclo
## continue → salta iteración


````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````
