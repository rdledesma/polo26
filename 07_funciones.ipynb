## Idea principal

Una función es un bloque de código reutilizable que realiza una tarea específica.


````python

print("Hola")
print("Hola")
print("Hola")
````


El código anterior repite instrucciones.

Con funciones podemos evitar eso.


````python

def saludar():
    print("Hola")

saludar()
saludar()
saludar()
````

# ¿Qué es una función?


Una función:
- Tiene un nombre
- Puede recibir datos
- Puede devolver resultados

Sintaxis general:

```python
def nombre_funcion():
    # código
```


# Primera función


Veamos una función sencilla.


````python

def mensaje():
    print("Bienvenidos al curso de Python")

mensaje()
````

# Return


Hasta ahora nuestras funciones solo imprimían.

Pero muchas veces necesitamos que una función DEVUELVA un resultado.

Para eso usamos `return`.


````python

def sumar(a, b):
    return a + b

resultado = sumar(10, 5)

print(resultado)
````


## Diferencia entre print y return

- `print()` muestra información
- `return` devuelve información


````python

def cuadrado(numero):
    return numero ** 2

x = cuadrado(4)

print(x)
````

# Funciones y listas

````python

def promedio(lista):
    return sum(lista) / len(lista)

numeros = [10, 20, 30, 40]

print(promedio(numeros))
````

# Scope (alcance de variables)


Las variables creadas dentro de una función solo existen dentro de ella.


````python

def ejemplo():
    x = 10
    print(x)

ejemplo()

# print(x)  # Esto daría error
````


## Variables globales


````python

nombre = "Ruben"

def mostrar():
    print(nombre)

mostrar()
````

# Funciones que llaman funciones

````python

def sumar(a, b):
    return a + b

def promedio(a, b):
    return sumar(a, b) / 2

print(promedio(10, 20))
````

# Funciones Lambda


Las funciones lambda son funciones pequeñas escritas en una sola línea.


````python

cuadrado = lambda x: x ** 2

print(cuadrado(5))
````

````python

numeros = [1, 2, 3, 4, 5]

dobles = list(map(lambda x: x * 2, numeros))

print(dobles)
````
